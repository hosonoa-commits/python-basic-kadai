```python
class Human:
 def __init__(self,name,age):
  self.name = name
  self.age = age

 def check_adult(self):
  if self.age >= 20:
    print(f"{self.name}さんは{self.age}歳なので大人です。")
  else:
   print(f"{self.name}さんは{self.age}歳なので大人ではありません。")

people_data={"A":10,"B":15,"C":20,"D":25}

for name,age in people_data.items():
 person = Human(name,age)
 person.check_adult()
```
